```
#Practice>Python>Regex and Parsing>Validating Credit Card Numbers
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

n = int(input())
for _ in range(n):
    s = input()
    if len(s) == 16:
        if not re.search(r"[\d]", s):
            print("Invalid")
        elif not re.search(r"^[456]", s):
            print("Invalid")
        elif re.search(r"(\d)\1{3,}", s):
            print("Invalid")
        else:
            print("Valid")
    elif len(s) == 19:
        if not re.search(r"[\d-]", s):
            print("Invalid")
        elif not re.search(r"[\d]{4}-[\d]{4}-[\d]{4}-[\d]{4}", s):
            print("Invalid")
        elif not re.search(r"^[456]", s):
            print("Invalid")
        elif re.search(r"(\d)\1{3,}", s):
            print("Invalid")
        else:
            print("Valid")
    else:
        print("Invalid")
```

```
#Practice>Python>Python Functionals>Validating Email Addresses With a Filter
def fun(s):
    # return True if s is a valid email, else return False
    lst = ''.join('abcdefghijklmnopqrstuvwxyz0123456789_-')
    b = False
    try:
        l = s.split('@')
        m = l[1].split('.')
    except:
        return False
    if(len(l)==2 and len(m)==2 and len(m[1])<4):
        if(m[0].isalpha() or m[0].isnumeric() or m[0].isalnum()):
            for letters in l[0]:
                if(letters in lst):
                    b = True
                else:
                    b = False
                    break
    return b
```

```
#Practice>Algorithms>Recursion>Recursive Digit Sum
#!/bin/python3

import math
import os
import random
import re
import sys

def rec(n):
    if n<10:
        return n
    else:
        n = str(n)
        return rec(sum(int(i) for i in n))

# Complete the superDigit function below.
def superDigit(n, k):
    s = sum(int(i) for i in n) * k
    return rec(s)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nk = input().split()

    n = nk[0]

    k = int(nk[1])

    result = superDigit(n, k)

    fptr.write(str(result) + '\n')

    fptr.close()
```
